```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from itertools import accumulate
import math
```

```python
monedas = 800 # placed the number of coins find
```

```python
def retorno_capitan(x):
    monedas_capitan = (monedas - (x * int(monedas / x)))
    return monedas_capitan
```

```python
monedas_for_captain = []
nro_of_coins_for_pirates = []
for i in range(1,monedas-1):
    vals = retorno_capitan(i)
    monedas_for_captain.append(vals)
    nroofcoins = i
    nro_of_coins_for_pirates.append(nroofcoins)
```

```python
d = {'nro_of_coins_for_pirates' : nro_of_coins_for_pirates , 'monedas_for_captain': monedas_for_captain}
data = pd.DataFrame(data=d)
```

```python
from matplotlib.ticker import MultipleLocator
minorLocator   = MultipleLocator(10)
#ax.xaxis.set_minor_locator(minorLocator)
#plt.grid(which='minor')
```

```python
print('Maximium coins gained by captain : ' + str(data['monedas_for_captain'].max()))
fig , ax1 = plt.subplots(1,1 , figsize=(15,5))
data.plot.bar(x='nro_of_coins_for_pirates', y='monedas_for_captain', rot=0 , ax=ax1)
ax1.set_xlabel('Number of pirates send')
ax1.set_ylabel('Number of coins for the captain')
ax1.tick_params(axis='x', direction='out', length=1, width=1, grid_alpha=0.5)
#ax1.tick_params(axis='x', which='major', pad=20)
```

Function will always show 0 coins for the captain whenever the value of number of coins divided the number of pirates is an integer , he will always give all the coins to his pirates. Function will always be like a "concave" function where we can find a global maximium
